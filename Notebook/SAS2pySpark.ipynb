```python
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql.functions  import *

data= [
       ["Upen-dra"],
       ["Am-it"],
       ["Ku-mar"],
       ["HS-BC"],
       ["Ku-nal"]
       ]

columns=["cin_sfx"]

spark = SparkSession.builder.appName('sas2py').getOrCreate()
barbell=spark.createDataFrame(data=data,schema=columns)
barbell.createOrReplaceTempView("barbell")
barbell.show()
```

```python
from pyspark.sql.functions import regexp_replace
barbell_list = spark.sql("""
select distinct trim(regexp_replace(cin_sfx,"-","")) as newcinsfx from barbell
""")
barbell_list.createOrReplaceTempView("barbell_list")
barbell_list.show()
```
